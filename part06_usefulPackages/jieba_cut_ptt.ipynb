```python
import os
import jieba
```

```python
source_path = r'./res_gossiping'
file_list = os.listdir(source_path)
file_list
```

```python
article_str = ''
for article in file_list:
    with open(source_path + '/' + article, 'r', encoding='utf-8') as f:
        article_str += f.read().split('---split---')[0]
#         article_str += tem_str
article_str
```

```python
new_str = article_str.replace('\n', ' ')
s = jieba.cut(new_str)

word_count = {}
for w in '|'.join(s).split('|'):
    if w in word_count:
        word_count[w] += 1
    else:
        word_count[w] = 1
word_count
```

```python
word_count_list = []
for wc in word_count:
    word_count_list.append((wc, word_count[wc]))
word_count_list
```

```python
word_count_list.sort(key = lambda x:x[1], reverse = True)
word_count_list
```
