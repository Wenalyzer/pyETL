```python
# 引入必要套件
import pandas as pd
import os

# 列出所有文檔
source_file_path = r'../part2_pttArticleWithCookie/res_gossiping'
file_list = os.listdir(source_file_path)
file_list
```

```python
# 先宣告一個DataFrame，並制定欄位名稱
df = pd.DataFrame(columns = ['文章編號', '推數', '噓數', '文章分數', '文章作者', '文章名稱','發表時間'])
df
```

```python
# 讀取每一篇文章，並將需要的部分取出，轉成我們需要個格式
article_info_row_data = []
for n, each_article in enumerate(file_list):
    article_path = source_file_path + '/' + each_article
    with open(article_path, 'r', encoding = 'utf-8') as f:
        tmp_article_string = f.read()
    tmp_row_data = [str(n+1)] + tmp_article_string.split('---split---')[1].split('\n')[1:-1]
    article_info_row_data.append(tmp_row_data)
article_info_row_data
```

```python
# 試著插入一筆資料
df.append(pd.DataFrame([['1',
  '推: 0',
  '噓: 0',
  '分數: 0',
  '作者: caeasonfb (eason)',
  '標題: Re: [問卦] 台灣如果沒有冷氣是不是無法現代化?',
  '時間: Sat Sep  7 16:47:10 2019']], columns = ['文章編號', '推數', '噓數', '文章分數', '文章作者', '文章名稱','發表時間']))
```

```python
# 將所有資料插入
df.append(pd.DataFrame(article_info_row_data, columns = ['文章編號', '推數', '噓數', '文章分數', '文章作者', '文章名稱','發表時間']))
```

```python
# 儲存成csv檔
df = df.append(pd.DataFrame(article_info_row_data, columns = ['文章編號', '推數', '噓數', '文章分數', '文章作者', '文章名稱','發表時間']))
df.to_csv(r'./pttGossiping.csv', index=0, encoding = 'utf-8')
```

```python
# 製作過濾器
def column_filter(s):
    return s.split(':')[-1].strip()
```

```python
df['推數'].apply(column_filter)
```

```python
for c in df:
    df[c] = df[c].apply(column_filter)
df
```
